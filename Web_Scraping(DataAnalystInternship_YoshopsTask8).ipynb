***Scraping the Data of Aurangabad Schools***

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
# Function to extract the school name
def get_name(soup):
  try:
    name = soup.h1.text
  except AttributeError:
   name = ""
  return name
# Function to extract year of establishment
def get_year(soup):
  try:
    year_of_establishment = soup.find('div',class_="schoolInformation").text
  except AttributeError:
    year_of_establishment = ""
  return year_of_establishment
# Function to extract offered streams
def get_stream(soup):
  try:
    offered_streams = soup.find('h5',class_="m-0").text
  except AttributeError:
    offered_streams = ""
  return offered_streams
# Function to extract no of teachers
def get_teacher(soup):
  try:
    no_of_teacher = soup.find('div', class_="col mb-3").text
  except AttributeError:
    no_of_teacher = ""
  return no_of_teacher
# Function to extract address of the school
def get_address(soup):
  try:
    address = soup.find('div',class_="d-flex align-items-top contact_content").text
  except AttributeError:
    address = ""
  return address
```

```python
links_list = []

HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36','Accept-Language': 'en-US, en;q=0.5'})

# URL of the pages of the website
URL = ['https://school.careers360.com/schools/schools-in-aurangabad','https://school.careers360.com/schools/schools-in-aurangabad?page=2&state=15%2C4&sort_by=1','https://school.careers360.com/schools/schools-in-aurangabad?page=3&state=15%2C4&sort_by=1']



# Extract all a tag from the pages
for url in range(URL.index(URL[-1])+1):
  webpage = requests.get(URL[url], headers=HEADERS)
  soup = BeautifulSoup(webpage.content,"html.parser")
  for data in soup.find_all('div', class_="title"):
    for a in data.find_all('a'):
        print(a.get('href')) #for getting link
        links_list.append(a.get('href'))
```

```python
d = {"Name":[], "Established":[],"Offered_streams/offered_courses":[],"No_of_teacher/No_of_faculty":[],"Address":[]}
```

```python
for link in links_list:
  new_webpage = requests.get(link, headers=HEADERS)
  new_soup = BeautifulSoup(new_webpage.content, "html.parser")
  d['Name'].append(get_name(new_soup))
  d['Established'].append(get_year(new_soup))
  d['Offered_streams/offered_courses'].append(get_stream(new_soup))
  d['No_of_teacher/No_of_faculty'].append(get_teacher(new_soup))
  d['Address'].append(get_address(new_soup))
```

```python
df = pd.DataFrame.from_dict(d)
```

```python
df.to_csv('AurangabadSchool.csv')
```

***Scraping the Data of Aurangabad Colleges***

```python
# Function to extract the school name
def get_name(soup):
  try:
    name = soup.h1.text
  except AttributeError:
   name = ""
  return name
# Function to extract year of establishment
def get_year(soup):
  try:
    results = soup.find('div',class_="tupple highlight")
    established = []
    for result in results:
     for row in result.find_all('tr'):
      if ('Established' in row.text):
        established.append(row.text)
  except AttributeError:
    established = ""
  return established
# Function to extract offered streams
def get_courses(soup):
  try:
    offered_courses = soup.find('div',class_="tagsBlk").text
  except AttributeError:
    offered_courses = ""
  return offered_courses
# Function to extract no of faculty
def get_faculty(soup):
  try:
    faculty = []
    results = soup.find('div',class_="tupple highlight")
    for result in results:
     for row in result.find_all('tr'):
      if ('Faculty' in row.text):
        faculty.append(row.text)
  except AttributeError:
    faculty = ""
  return faculty
# Function to extract address of the school
def get_address(soup):
  try:
    address = soup.find_all('div',class_="contct col-sm-5")[0].text
  except AttributeError:
    address = ""
  return address
```

```python
links_list = []

HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36','Accept-Language': 'en-US, en;q=0.5'})

# URL of the pages of the website
URL = ['https://www.careers360.com/colleges/aurangabad-colleges-fctp','https://www.careers360.com/colleges/aurangabad-colleges-fctp?page=2&sort_by=3&state=15%2C4%2C8%2C29','https://www.careers360.com/colleges/aurangabad-colleges-fctp?page=3&sort_by=3&state=15%2C4%2C8%2C29','https://www.careers360.com/colleges/aurangabad-colleges-fctp?page=4&sort_by=3&state=15%2C4%2C8%2C29']



# Extract all a tag from the pages
for url in range(URL.index(URL[-1])+1):
  webpage = requests.get(URL[url], headers=HEADERS)
  soup = BeautifulSoup(webpage.content,"html.parser")
  for data in soup.find_all('h3', class_="college_name"):
    for a in data.find_all('a'):
        print(a.get('href')) #for getting link
        links_list.append(a.get('href'))
```

```python
d = {"Name":[], "Established":[],"Offered_streams/offered_courses":[],"No_of_teacher/No_of_faculty":[],"Address":[]}
```

```python
for link in links_list:
  new_webpage = requests.get(link, headers=HEADERS)
  new_soup = BeautifulSoup(new_webpage.content, "html.parser")
  d['Name'].append(get_name(new_soup))
  d['Established'].append(get_year(new_soup))
  d['Offered_streams/offered_courses'].append(get_courses(new_soup))
  d['No_of_teacher/No_of_faculty'].append(get_faculty(new_soup))
  d['Address'].append(get_address(new_soup))
```

```python
df = pd.DataFrame.from_dict(d)
```
