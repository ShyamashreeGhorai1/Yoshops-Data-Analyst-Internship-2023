```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
def get_name(soup):
  try:
    name = soup.h1.text
  except AttributeError:
    name = ''
  return name
def get_price(soup):
  try:
    price = soup.find_all('div',class_="_30jeq3 _16Jk6d")[0].text
  except AttributeError:
    price = ''
  return price
def get_rating(soup):
  try:
    rating = soup.find_all('div',class_="_3LWZlK")[0].text
  except AttributeError:
    rating = ''
  return rating
def get_type(soup):
  try:
    descriptions = soup.find_all('li',class_="_21Ahn-")
    Type = []
    for i in range(len(descriptions)):
      q = descriptions[i].text
      if ('Type' in q):
        Type.append(q)
  except AttributeError:
    Type = ''
  return Type
def get_material(soup):
  try:
    descriptions = soup.find_all('li',class_="_21Ahn-")
    Material = []
    for i in range(len(descriptions)):
      q = descriptions[i].text
      if ('Material' in q):
        Material.append(q)
        return Material
  except AttributeError:
    Material = ''
  return Material
def get_color(soup):
  try:
    descriptions = soup.find_all('li',class_="_21Ahn-")
    Color = []
    for i in range(len(descriptions)):
      q = descriptions[i].text
      if ('Color' in q):
        Color.append(q)
        return Color
  except AttributeError:
    Color = ''
  return Color
def get_quantity(soup):
  try:
    descriptions = soup.find_all('li',class_="_21Ahn-")
    Quantity = []
    for i in range(len(descriptions)):
      q = descriptions[i].text
      if ('Quantity' in q):
        Quantity.append(q)
        return Quantity
  except AttributeError:
    Quantity = ''
  return Quantity
def get_width(soup):
  try:
    descriptions = soup.find_all('li',class_="_21Ahn-")
    Width = []
    for i in range(len(descriptions)):
      q = descriptions[i].text
      if ('Width' in q):
        Width.append(q)
        return Width
  except AttributeError:
    Width = ''
  return Width
def get_ideal_for(soup):
  try:
    descriptions = soup.find_all('li',class_="_21Ahn-")
    Ideal_For = []
    for i in range(len(descriptions)):
      q = descriptions[i].text
      if ('Ideal For' in q):
        Ideal_For.append(q)
        return Ideal_For
  except AttributeError:
    Ideal_For = ''
  return Ideal_For
```

```python

URL = ['https://www.flipkart.com/search?q=HRX+price+less+than+500&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY','https://www.flipkart.com/search?q=HRX+price+less+than+500&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY&page=2','https://www.flipkart.com/search?q=HRX+price+less+than+500&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY&page=3','https://www.flipkart.com/search?q=HRX+price+less+than+500&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY&page=4']
links = []
start_link = "https://www.flipkart.com"
for url in range(URL.index(URL[-1])+1):
  webpage = requests.get(URL[url])
  soup = BeautifulSoup(webpage.content,"html.parser")
  data = soup.find_all('div',{'class':"_1xHGtK _373qXS"})
  for items in data:
    rest_link = items.find('a')['href']
    links.append(start_link+rest_link)
```

```python
d = {"Name":[], "Price":[],"Rating":[],"Type":[],"Material":[],"Color":[],"Quantity":[],"Width":[],"Ideal_For":[]}
```

```python
for link in links:
  new_webpage = requests.get(link)
  new_soup = BeautifulSoup(new_webpage.content, "html.parser")
  d['Name'].append(get_name(new_soup))
  d['Price'].append(get_price(new_soup))
  d['Rating'].append(get_rating(new_soup))
  d['Type'].append(get_type(new_soup))
  d['Material'].append(get_material(new_soup))
  d['Color'].append(get_color(new_soup))
  d['Quantity'].append(get_quantity(new_soup))
  d['Width'].append(get_width(new_soup))
  d['Ideal_For'].append(get_ideal_for(new_soup))
```

```python
df = pd.DataFrame.from_dict(d)
```
